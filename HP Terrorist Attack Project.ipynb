```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset
file_path='terrorist-attacks.csv'
df=pd.read_csv(file_path)

# Creating a function
def terrorist_attacks_over_year(country_name):
    global india_df
    # Filter data for India
    india_df = df[df['Entity']==country_name.title()]
    
    # EDA for Terrorist Attacks in India
    
    # Plot 1: Line Plot-Terrorist Attacks Over the Years
    plt.figure(figsize=(12,6))
    plt.plot(india_df['Year'].to_numpy(), india_df['Terrorist attacks'].to_numpy(), marker='o', linestyle='-', color='b')
    plt.title(f'Terrorist Attacks in {country_name} Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Number of Terrorist Attacks')
    plt.grid(True)
    plt.show()
# Insight 1: The line plot reveals the overall trend in terrorist attacks in India over the years. We can identify periods with increased or decreased activity.
terrorist_attacks_over_year('India')
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
file_path = 'terrorist-attacks.csv'
df = pd.read_csv(file_path)

import warnings
warnings.filterwarnings('ignore')

def terrorist_attacks_over_years(country_name):
    global india_df
    # Filter data for India
    india_df = df[df['Entity'] == country_name.title()]

    # EDA for Terrorist Attacks in India

    # Plot 1: Line Plot - Terrorist Attacks Over the Years
    plt.figure(figsize=(12, 6))
    plt.plot(india_df['Year'].to_numpy(), india_df['Terrorist attacks'].to_numpy(), marker='o', linestyle='-', color='b')
    plt.title(f'Terrorist Attacks in {country_name} Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Number of Terrorist Attacks')
    plt.grid(True)

    # Fit ARIMA model and display forecast
    country_df, country_model, country_forecast = fit_arima_model(country_name)
    print(f"Forecast for {country_name}: {country_forecast}")
    
    # Plot the forecast with a different color
    years_forecasted = np.arange(india_df['Year'].max() + 1, india_df['Year'].max() + 11)  # Ensure continuous x-axis values
    plt.plot(years_forecasted, country_forecast.to_numpy(), marker='o', linestyle='--', color='r', label='Forecast')
    plt.legend()
    
    plt.show()

# Insight 1: The line plot reveals the overall trend in terrorist attacks in India over the years. We can identify periods with increased or decreased activity.

def fit_arima_model(country):
    country_df = df[df['Entity'] == country]
    model = ARIMA(country_df['Terrorist attacks'], order=(5,1,0))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=10)
    return country_df, model_fit, forecast

# Plot terrorist attacks over years for Afghanistan
terrorist_attacks_over_years('Ukraine')
```

## Importing Necesary Libraries

```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
```

## Intialize H2O Cluster

```python
h2o.init()
```

## Load the Dataset

```python
df=pd.read_csv('terrorist-attacks.csv')
df
```

## Convert Pandas DataFrame to H2O Frame

```python
h2o_df = h2o.H2OFrame(df)
```

## Creating a function for death ages

```python
def calculate_death_range(df, entity_column, year_column, columns):
    # Calculate total death range for each row
    df['Death Range'] = df[columns].sum(axis=1)
    
    # Select the columns and drop duplicates
    result = df[[entity_column, year_column, 'Death Range']].drop_duplicates()
    return result

# Define the columns cntaining death range
people_death = ['Death Age 100+', 'Death Age: 51-99 ', 'Death Age : 21-50 ', 'Death Age : 11-20 ', 'Death Age : 6-10 ', 'Death Age :  1-5']

# Calculate total death range for each entity and year
total_death_range = calculate_death_range(df, 'Entity', 'Year', people_death)
total_death_range.head()
```

```python
# Merging the datset from groupby
df['Death Range'] == total_death_range['Death Range']
df.sample(10)
```

## Identify predictors and response

```python
x = h2o_df.columns
y = 'Death Range'
x.remove(y)
```

## Run AutoML

```python
aml = H2OAutoML(max_runtime_secs=600)
# You can adjust the max runtime as needed
aml.train(x=x, y=y, training_frame=h2o_df)
```
